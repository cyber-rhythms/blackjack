### Investigation of autocorrelation plots

Let's investigate the autocorrelation plots of the time-series. Strategy will be to hand-code these functions, and check their consistency against statsmodels.

* For the autocorrelation function, hand-coded functions give the same results as statsmodels acf plot, verified by examination of plots, but not by printing any arrays. Verified for 30 and 250 autocorrelations. And also numerically against statmodel output.

* Yule-Walker method - create a k x k array of sample autocorrelation coefficients, take its inverse and premultiply by a dimension-k vector of sample autocorrelation coefficients.

* Need to QUICKLY CHECK for consistency of both against statmodels pacf arrays. Currently a long way off on the pacf front, but nothing that can't be fixed with some tinkering.

```python
# Define a forward operator function

def apply_forward_operator(series, lag):
    
    "Takes a series and repeatedly applies the forward operator to it, with number of operations determined by lag"
    
    nan_array = np.array([np.nan for index in range(lag)])
    indices = [index for index in range(lag)]
    forward_series = np.delete(series, indices)
    forward_series = np.append(forward_series, nan_array)
    
    return forward_series
```

```python
# Define the autocorrelation coefficient function

def autocorrelation_coeff(series, lag, sample_mean, sample_variance, total_observations):
    
    """Takes a series and lag length as input and returns a sample estimate of the kth lag autocorrelation coefficient."""
    
    forward_series = apply_forward_operator(series, lag)
    forward_series_diff = forward_series - sample_mean
    series_diff = series - sample_mean
    autocovariance = np.sum(forward_series_diff[0:total_observations - lag] * series_diff[0:total_observations - lag]) / total_observations 
    autocorrelation_coeff = autocovariance / sample_variance

    return autocorrelation_coeff
```

```python
# Define the autocorrelation function that takes lag length and returns
# all autocorrelation coefficients up to and including thtat lag length.

def autocorrelation_function(series, lags, sample_mean, sample_variance, total_observations):
    
    """Takes a series and lag as input and computes the autocorrelation function, returning an array of sample estimates of 
    autocorrelation coefficients from zero until the specified lag."""
    
    autocorrelation_array = np.array([[autocorrelation_coeff(series, lag, sample_mean, sample_variance, total_observations) for lag in range(lags)]])
    
    return autocorrelation_array
```

```python
# Compute partial autocorrelations, beginning with 30 lags, but in principle for any number of lags and validate against
# statsmodels

def autocorrelation_matrix(series, lags, sample_mean, sample_variance, total_observations):
    
    sample_autocorrelation_coefficient = autocorrelation_coeff(series, 0, sample_mean, sample_variance, total_observations)
    sample_autocorrelation_matrix = np.diagflat([sample_autocorrelation_coefficient for i in range(lags)])
    for diagonal in range(1, lags):
        sample_autocorrelation_coefficient = autocorrelation_coeff(series, diagonal, sample_mean, sample_variance, total_observations)
        sup_diagonals = np.diagflat([sample_autocorrelation_coefficient for i in reversed(range(lags - diagonal))], diagonal)
        sub_diagonals = np.diagflat([sample_autocorrelation_coefficient for i in reversed(range(lags - diagonal))], -diagonal)
        sample_autocorrelation_matrix += (sup_diagonals + sub_diagonals)
        
    return sample_autocorrelation_matrix 
```

```python
def partial_autocorrelation_function(series, lags, sample_mean, sample_variance, total_observations):
    sample_autocorrelation_matrix = autocorrelation_matrix(series, lags, sample_mean, sample_variance, total_observations)
    sample_autocorrelation_vector = np.array([autocorrelation_coeff(series, lag, sample_mean, sample_variance, total_observations) for lag in range(1, lags + 1)])
    
    partial_autocorrelation = (np.linalg.inv(sample_autocorrelation_matrix)).dot(sample_autocorrelation_vector)
    
    return sample_autocorrelation_matrix, sample_autocorrelation_vector, partial_autocorrelation
```

### Formalisms

* Stochastic process, time series as realisation, 

* Assume that each observation $x_t$ is a realisation of a random variable $X_t$.

* A *stochastic process* is a family of random variables, $\left\{X_1, X_2, X_3,..., X_T\right\}$, often abbreviated as $\left\{X_t, t \in T\right\}$, defined on a probability space $(\Omega, \mathscr{F}, P)$.

* A time-series can be thought of as a *realisation* of a stochastic process, and is denoted by $\left\{x_t, t \in T_0 \right\}$.

* $T \supseteq T_0$

* The functions $\left\{X(\omega), \omega \in \Omega \right\}$ on $T$ are known as the realisations or sample-paths of the process $\left\{X_t, t \in T\right\}$. 

* We will assume in our context that time is discrete.

* For a finite number of random variables, autocovariance matrices can be used computed to gain insight into dependence.

* For a time series $\left\{X_t, t \in T \right\}$, the autocovariance matrices need to be extended to deal with potentially infinite collections of random variables.

### White-box specification of the model

Denote the player's initial capital stock in period 0 as $K_0$.

In the next period, their capital stock will be:

$$K_1 = K_0 + u_1$$

In the period after, their capital stock will be:

$$K_2 = K_1 + u_2$$

That is, the capital stock in each round is the player's previous capital stock, plus a current round shock or innovation variable, $u_t$.

In general then, the player capital stock in the next round can be specified via the recursive relation:

$$K_{t+1} = K_t + u_{t+1} \quad t = 0,..., T$$

Because we have knowledge of the process, under fixed betting behaviour, we can give a specify the sample space of this innovation $u_t$, that is the set of all possible outcomes it can take.

We can make successive substitutions 

Define the player's bet per round as $b$, as this is under fixed betting behaviour. Then $u_t$ is a random variable with sample space $\Omega = \{-2, -1, 0, 1, 1.5, 2\}$.

$$  u_t =
\begin{cases}
2b,  & \text{if player doubles and wins the hand} \\
1.5b, & \text{if player blackjacks} \\
b, & \text{if player wins the hand} \\
0b, & \text{if the outcome of the hand is a draw} \\
-b, & \text{if the player loses the hand} \\
-2b & \text{if the player doubles and loses the hand} \\
\end{cases}$$
